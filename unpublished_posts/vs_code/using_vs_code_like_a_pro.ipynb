# Enable MCP

* Set `chat.mcp.access` to `all` in your `settings.json`.
* Set `chat.mcp.gallery.enabled` to `true` in your `settings.json`.
* Install MCP servers from the gallery (same window as the plugins).

## Helpful MCP Servers 

The MCP server settings are available by clicking the gear icon next to the MCP server in the gallery. This same file with contain the API keys obtained from these MCP servers (fully managed by VS Code). 

* [Context7](https://github.com/mcp/upstash/context7) - Provides up-to-date documentation for any library/framework.
* [Markitdown](https://github.com/mcp/microsoft/markitdown) - Converts between Markdown/HTML/PDF, etc.
* [Github](https://github.com/mcp/github/github-mcp-server) - Interact with GitHub repositories.
* [Chrome Dev Tools](https://github.com/mcp/chromedevtools/chrome-devtools-mcp) - Interact with Chrome Dev Tools in am active Chrome window via MCP. 
* [Stripe](https://github.com/mcp/stripe/agent-toolkit) - Interact with Stripe API via MCP.
* [Figma](https://github.com/mcp/figma/mcp-server-guide) - Turn Figma designs into code via MCP.
* [Desktop Commander](https://example.com) - Work with files and assets. image resize, split big files, refactor, and even do a git commit after it's done...

## How to Call MCP Servers

Make sure you're in Agent mode (not Chat mode).

You can use `@context7`, `@markitdown`, etc. to call the respective MCP servers.
