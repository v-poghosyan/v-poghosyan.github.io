# No BS Guide to Setting Up Your First Local NAS (Network Attached Storage) for Your Digital Garden

This can be useful to link to local files in your digital garden (such as tutorial videos), and also to have a local backup of your digital garden (in case Github or other services go down). It can also be useful for hosting a local instance of your digital garden, if you want to access it from other devices on your local network -- so you don't have to rely on quarto preview for the unpublished stuff (although that's certainly a workflow that's been working so far).

But, the main advantage of a NAS is that I can access these videos and other files from anywhere. Meaning, I can take out my laptop (which is no longer bogged down with all those video files) and through a local build of this digital garden (this server-side rendered Quarto garden), I can view my videos. So, essentially I have the holy grail of personal operating systems -- a note-augmented file system that I can access from anywhere.

Another benefit of NAS is that I can run Git Server on it, and move these unpublished notes (not intended for public consumption) off Microsoft's GitHub.

In this tutorial, we will be setting up a Synology NAS.

# How to link to files on your NAS from your digital garden?

Simply link to `http://<your-nas-ip-address>/web/path/to/your/file.mp4`. Remember, this IP is global, so you can access it from anywhere. All you need to do is remember this address, avoid creating a domain name for it.
