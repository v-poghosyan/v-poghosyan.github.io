# n8n Automations

[n8n](https://n8n.io/) is an open-source workflow automation tool that allows you to connect various applications and services to automate tasks and processes. It provides a visual interface where you can create workflows by connecting different nodes, each representing a specific action or trigger.

# APIs

With n8n you can hit any API with an API node. You can use the HTTP Request node to make API calls to any service that provides a RESTful API. You can configure the node to specify the HTTP method (`GET`, `POST`, `PUT`, `DELETE`, etc.), the URL of the API endpoint, headers (which can include authorization type bearer tokens for authentication), query parameters (which can include api-tokens for authentication), and request body.

# Webhooks

Webhooks are the inverse of APIs ($API^{-1}$). We can use the Webhook node to *create* endpoints (rather than hit them), and rather than reaching out to an endpoint we can set up an endpoint that receives data from external services. When a request is made to the webhook URL, it triggers the workflow to start executing. This is useful for integrating with services that can send data via webhooks, such as form submissions, payment notifications via Stripe, or other event-driven actions.
