# Navigate

Symbol `~` is your home directory.

```sh
cd ~
``` 

To go back one directory:

```sh
cd ..
```

To go back to the previous directory:

```sh
cd -
```

# List files

```sh
ls -l
```

Optionally, use `-a` flag to show hidden files:


# Download anything

Right-click on hyperlinks and click "copy address," then use `wget` command:

```sh
wget https://example.com/file
```

# Unzip a file

If the file is a zip file, use `unzip` command:

```sh
unzip file.zip
```

# Find a specific file

To find a specific file, use `find` command. Suppose we are looking for `secret-text.txt` file:

```sh
find . -name secret-text.txt
```

This command searches for `secret-text.txt` in the current directory (`.`) and all its subdirectories.

# Examine file

Examine type:

```sh
file file
```

Examine contents:

```sh
cat file
```

# Find a substring in file 

```sh
egrep "substring" file
```

Difference between `grep` and `egrep` is that `egrep` supports [extended regular expressions](https://en.wikibooks.org/wiki/Regular_Expressions/POSIX-Extended_Regular_Expressions).

# Search command history

```sh
history | egrep "substring"
```

# Git commands 

Do this instead of just `git log` to see condensed commit history... 

```sh
git log origin/main --oneline | head
```
